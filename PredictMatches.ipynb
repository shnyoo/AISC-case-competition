We created a dataset with dummy values of over 200 students.

```python
import pandas as pd
import random
import numpy as np
from sklearn.preprocessing import LabelEncoder


df = pd.read_csv('MentorMatch.tsv', index_col=None , delimiter = '\t')
df
```

Encoding the fields


```python
# Define your categorical columns
categorical_columns = ['City', 'Gender', 'Academic Standing', 'Type of Admission', 'Major', 'Race', 'Academic Interest 1', 'Academic Interest 2', 'Academic Interest 3']

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Iterate over each categorical column and transform it
for column in categorical_columns:
    encoded_column = label_encoder.fit_transform(df[column])  # Assuming 'df' is your DataFrame and 'column' is the categorical column
    df[column] = encoded_column
```

```python
Major = df['Major'].unique()
Names = df['Name'].unique()
Ages = df['Age'].unique()
Cities = df['City'].unique()
Gender = df['Gender'].unique()
Academic_Standing = df['Academic Standing'].unique()
Type_of_Admission = df['Type of Admission'].unique()
Race = df['Race'].unique()
Academic_Interest_1 = df['Academic Interest 1'].unique()
Academic_Interest_2 = df['Academic Interest 2'].unique()
Academic_Interest_3 = df['Academic Interest 3'].unique()
```

```python
column_unique = {'Major': [], 'Name': [], 'Age': [], 'City': [], 'Gender': [], 'Academic Standing': [], 'Type of Admission': [], 'Race': [], 'Academic Interest 1': [], 'Academic Interest 2': [], 'Academic Interest 3': []}

for k, lst in zip(column_unique.keys(), [Major, Names, Ages, Cities, Gender, Academic_Standing, Type_of_Admission, Race, Academic_Interest_1, Academic_Interest_2, Academic_Interest_3]):
    column_unique[k].extend(lst)
    # print(k)

# print(column_unique['Major'])
```

```python
# Creating a Dataset of men and women
men = pd.DataFrame()

women = pd.DataFrame()

# Number of users
num = 1000

# Dating profile questions for each
qs = ['Major', 'Type of Admission', 'Gender', 'City', 'Academic Interest 1', 'Academic Interest 2', 'Academic Interest 3']

# Answers to profile questions
# ans = {}
# ans['Major'] = column_unique['Major']

for q in qs:
    ans = column_unique[q]
    # Making them categorical for preprocessing later
    men[q] = pd.Categorical(random.choices(ans, k=num), categories=ans)
    
    women[q] = pd.Categorical(random.choices(ans, k=num), categories=ans)
    
    # IDs
    men['id'] = ["m"+str(i) for i in range(num)]
    
    women['id'] = ["mentor"+str(i) for i in range(num)]
    
# Setting index
men.set_index('id', inplace=True)

women.set_index('id', inplace=True)

# Creating match status between users
ratings = pd.DataFrame(index=men.index, columns=women.index)

for i in ratings.columns:
    ratings[i] = random.choices([0,1,"unseen"], k=num)
```

Creating a dummy user interaction table between our 200 users, which is swiping left (no), swiping right (yes), or unseen (no interaction between the users)

```python
def matchMan(men_df, women_df, ratings, new_man_answers, num_sim=10):
    """
    This function will return the most likely compatible women based on a few given
    dataframes for a new male user.  Will use the top N similar users' compatibility 
    ratings to find the potentially most compatible women.
    """
    
    # First need to replace the DF answers with their numerical values
    men_df = men_df.apply(lambda x: x.cat.codes)

    women_df = women_df.apply(lambda x: x.cat.codes)
    # print("new_ans", len(new_man_answers))
    # new_man_answers = new_man_answers.reshape(-1, 1)
    # print("cols", men_df.columns)
    
    # Dataframe of new user
    new_man = pd.DataFrame(
        [new_man_answers],
        columns=men_df.columns,
        index=['mentor'+str(int(men_df.index[-1][1:])+1)] # Getting the new man's id 
    )

    # print("new_man", new_man)
    
    # Categorical answers to the profile questions
    # ans = ['A', 'B', 'C', 'D', 'E']
    
    # Categorizing the answers
    new_man = new_man.apply(
        lambda x: pd.Categorical(x, categories=ans)
    ).apply(
        lambda x: x.cat.codes, axis=1
    )
        
    # Getting the top N similar users
    sim_men = men_df.corrwith(
        new_man.iloc[0], 
        axis=1
    ).sort_values(ascending=False)[:num_sim].index
    
    # Getting the similar users' ratings
    sim_rate = ratings.T[sim_men]
    
    # Filling in unseen values with nan for calculation purposes
    sim_rate.replace("unseen", np.nan, inplace=True)
    
    # The potentially most compatible women for the new man
    most_comp = sim_rate.mean(axis=1).sort_values(ascending=False)
    
    return most_comp[10:]
```

```python
new_man_answers = []
new_man_answers = random.choices(Major, k = 1)
ac_1 = random.choices(Academic_Interest_1)
ac_2 = random.choices(Academic_Interest_2)
ac_3 = random.choices(Academic_Interest_3)
city = random.choices(Cities)
# new_man_answers.append(random_major)
type_of_admission = 1
ages = random.choices(Ages)
race = random.choices(Race)
new_man_answers.append(type_of_admission)
gender = 1
new_man_answers.append(gender)
new_man_answers.append(city[0])
# new_man_answers.append(ages[0])
# new_man_answers.append(race[0])
new_man_answers.append(ac_1[0])
new_man_answers.append(ac_2[0])
new_man_answers.append(ac_3[0])
# new_man_answers = [random_major, type_of_admission, gender]
# Randomly picking answers
# new_man_answers = random.choices(ans, k=3)

print("New mentee details: \nMajor, Type Of Admission, Gender, City, Top 3 Academic Interests\n", new_man_answers)
```

```python

# Running the function
recs = matchMan(
    men, 
    women, 
    ratings, 
    new_man_answers, 
    num_sim=10
)
print("Top 20 Mentor Matches personalized for you!")
# Finding the top 20 most potentially compatible
recs[:20]
```
